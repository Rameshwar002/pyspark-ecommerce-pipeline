```python
df=pd.read_csv("d:\\GitAction\\pyspark-ecommerce-pipeline\\output\\transformed\enriched_data.csv")
print(df.columns.tolist())
df.columns = df.columns.str.strip()   # removes leading/trailing spaces


# Group by product_id and sum amounts
product_sales = df.groupby("product_id")["amount"].sum().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(10,6))
product_sales.plot(kind="bar", color="skyblue", edgecolor="black")

plt.title("Sales Amount by Product", fontsize=14)
plt.xlabel("Product ID", fontsize=12)
plt.ylabel("Total Sales Amount", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Take top 10 products
top_products = product_sales.sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
top_products.plot(kind="bar", color="skyblue", edgecolor="black")

plt.title("Top 10 Products by Sales Amount")
plt.xlabel("Product ID")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
top_products = product_sales.sort_values(ascending=False).head(20)

plt.figure(figsize=(10, 8))
top_products.plot(kind="barh", color="lightgreen", edgecolor="black")

plt.title("Top 20 Products by Sales Amount")
plt.xlabel("Total Sales Amount")
plt.ylabel("Product ID")
plt.tight_layout()
plt.show()
```
